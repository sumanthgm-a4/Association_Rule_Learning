```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## **Data Preprocessing**

```python
dataset = pd.read_csv("Market_Basket_Optimisation.csv", header = None)

transactions = []
for i in range(7501):
    transactions.append([str(dataset.values[i, j]) for j in range(20)])
```

## **Training the Apriori model on the whole dataset**

```python
from apyori import apriori

rules = apriori(transactions = transactions, min_support = 3*7/7501, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

## **Visualizing the results (all the rules sorted by relevance)**

Displaying the first results coming directly from the output of the apriori()

```python
results = list(rules)
results
```

Putting the results into a well-organised Pandas DataFrame

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```
